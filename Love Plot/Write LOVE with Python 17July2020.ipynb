```python
# Create "L" using [y = 1/x]
x = np.array(range(100))
y = 1/x

df_L = {'x': x,'y': y }
```

```python


# Create the plot
plt.plot(x,y,label='EQ: y = 1/x')

# Add a title
plt.title('The Letter "L"')

# Add X and y Label
plt.xlabel('x axis')
plt.ylabel('y axis')

# Add a grid
plt.grid(alpha=.4,linestyle='--')

# Add a Legend
#plt.legend()

# Show the plot
#plt.show()
sns.lineplot(x="x", y="y", data=df_L)
```
